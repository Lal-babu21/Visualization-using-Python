```python
data_canada = pd.read_excel('canada.xlsx')
```

```python
data_canada.head()
```

```python
data_canada.drop(['AREA','REG','Type','Coverage','DEV'], axis =1 , inplace=True)
```

```python
data_canada.head()
```

```python
data_canada.rename(columns = {'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
```

```python
data_canada.head()
```

```python
all(isinstance(column,str) for column in data_canada.columns)
```

```python
data_canada.columns = list(map(str, data_canada.columns))
```

```python
all(isinstance(column,str) for column in data_canada.columns)
```

```python
data_canada.set_index('Country',inplace=True)
```

```python
data_canada.head()
```

```python
data_canada['Total'] = data_canada.sum(axis = 1)
```

```python
data_canada.head()
```

```python
sorted_data = data_canada.sort_values(['Total'], ascending = False, axis =0).head()
```

```python
years = list(map(str,range(1980,2014)))
```

```python
years
```

```python
top5 = sorted_data
```

```python
data_to_chart = top5[years].transpose()
```

```python
data_to_chart.plot(kind = 'area',
                   figsize = (12,4),
                  stacked = True,
                  rot = 90)

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Immigration Trend for Top 5 Countries')
```

```python
data_to_chart.plot(kind = 'bar',
                   figsize = (12,4),
                  stacked = True)

plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.title('Immigration Trend for Top 5 Countries')
```

```python
df_iceland = data_canada.loc['Iceland',years]
```

```python
df_CI = data_canada.loc[['China','India'], years].transpose()
```

```python
fig = plt.figure()

ax0 = fig.add_subplot(1,2,1) #add subplot 1 (1 row, 2 colmns, first plot)
ax1 = fig.add_subplot(1,2,2) #add subplot 2 (1 row, 2 colmns, second plot)

# Subplot 1: Box Plot
df_CI.plot(kind = 'box',color = 'blue', vert = False, figsize = (20,6), ax = ax0)
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')
ax0.set_title('Box Plot of Immigrants from China and India (1980-2013)')

#Subplot 2: Line Chart
df_CI.plot(kind = 'line',figsize = (20,6), ax = ax1)
ax1.set_xlabel('Years')
ax1.set_ylabel('Number of Immigrants')
ax1.set_title('Line chart of Immigrants from China and India (1980-2013)')
```
